Khoảng cách euclidean giữa 2 ảnh 

```python
def TinhHist(pathfilename):
  img = cv2.imread(pathfilename, 0)
  hist = cv2.calcHist([img], [0], None, [256], [0, 256])
  size = img.shape[0]*img.shape[1]
  hist = hist/size
  return hist
```

```python

from scipy.spatial import distance

eclid_d1 = distance.euclidean(TinhHist(path_filename_1), TinhHist(path_filename_2))
print("Eclidean distance between case 1 is: ", eclid_d1)

eclid_d2 = distance.euclidean(TinhHist(path_filename_1), TinhHist(path_filename_3))
print("Eclidean distance between case 2 is: ", eclid_d2)
```

Tính sự tương đồng cosis giữa 2 ảnh

```python

cosin_d1 = 1 - distance.cosine(TinhHist(path_filename_1), TinhHist(path_filename_2))
print("Cosine distance between case 1 is: ", cosin_d1)

cosin_d2 = 1 - distance.cosine(TinhHist(path_filename_1), TinhHist(path_filename_3))
print("Cosine distance between case 2 is: ", cosin_d2)
```
